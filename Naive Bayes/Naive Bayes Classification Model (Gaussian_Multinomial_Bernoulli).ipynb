# Naive Bayes Classification

Naive Bayes is a probabilistic machine learning algorithm based on Bayes' Theorem. It assumes that all features are independent given the class label. Despite this strong assumption, Naive Bayes often performs well on real-world classification tasks.

In this notebook, we will explore three variants of Naive Bayes:

1. **Gaussian Naive Bayes (GaussianNB)**
   Used for continuous numerical features that approximately follow a normal distribution.

2. **Multinomial Naive Bayes (MultinomialNB)**
   Commonly used for text classification with word-count features.

3. **Bernoulli Naive Bayes (BernoulliNB)**
   Designed for binary-valued features (0/1).

# Import Libraries

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.datasets import load_breast_cancer, make_classification
import numpy as np
```

# Function For Each Model
For each model, we will:

* Split the dataset into training and testing sets.
* Train the classifier.
* Generate predictions.
* Evaluate accuracy.
* Display the confusion matrix.
* Show TP, TN, FP, and FN values.
* Print the classification report.

```python
def run_model(X, y, test_size, model_type):

    classes, counts = np.unique(y, return_counts=True)
    
    X_train, X_test, y_train, y_test = train_test_split(
        X,
        y,
        test_size=test_size,
        random_state=42,
        stratify=y
    )
    
    # Model
    model = model_type()
    model.fit(X_train, y_train)

    # Predictions
    y_pred = model.predict(X_test)

    # Performance
    train_acc = model.score(X_train, y_train)
    test_acc = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, zero_division=0)
    cm = confusion_matrix(y_test, y_pred)

   
    
    print(f"\n============== {model_type.__name__} ==============\n")

     # Dataset Info
    print("\nDataset Information")
    print("-" * 30)
    print("Shape:", X.shape)
    print("Features:", X.shape[1])
    print("Classes:", classes)
    print("Class Distribution:", dict(zip(classes, counts)))

    print("\nTrain-Test Split")
    print("-" * 30)
    print(f"Training Samples: {X_train.shape[0]}")
    print(f"Testing Samples : {X_test.shape[0]}")

    print("\nPredictions:")
    print("-" * 30)
    print(y_pred) 
    print("\nPerformance")
    print("-" * 30)
    print("Train Accuracy:", round(train_acc, 4))
    print("Test Accuracy :", round(test_acc, 4))
    print("\nClassification Report:\n",report)
    print("Confusion Matrix:\n",cm)

    # Binary classification only
    if cm.shape == (2, 2):
        tn, fp, fn, tp = cm.ravel()
    
        print("\nBinary Classification Metrics")
        print("-" * 30)
        print(f"True Negative = {tn}")
        print(f"False Positive = {fp}")
        print(f"False Negative = {fn}")
        print(f"True Positive = {tp}")
    
    else:
        print("\nMulticlass Classification Detected")
        print("-" * 30)
    
        for i in range(cm.shape[0]):
            print(f"Class {i} vs Rest:")
            print(f"  TP (diagonal) = {cm[i, i]}")
            print(f"  FN = {cm[i].sum() - cm[i, i]}")
            print(f"  FP = {cm[:, i].sum() - cm[i, i]}")

    if model_type == GaussianNB:

        print("\nExtra Info:")
        print("-" * 30)
        print("\nClass Prior Probabilities P(y):\n")
        print(model.class_prior_)

        print("\nFeature Means per Class (first 5 features):\n")
        print(model.theta_[:, :5])

        print("\nFeature Variances per Class (first 5 features):\n")
        print(model.var_[:, :5])

    return model
```

# Gaussian Naive Bayes (GNB)
* Features are continuous numerical values.
* Assumes each feature follows a Gaussian (Normal) distribution within each class.
* Commonly used for medical and scientific datasets.

## Dataset: Breast Cancer Dataset

```python
data = load_breast_cancer()

X = data.data
y = data.target

run_model(
    X,
    y,
    test_size=0.25,
    model_type=GaussianNB
)
```

# Multinomial Naive Bayes (MNB)

* Works with count data.

* Frequently used in:

  * Spam detection
  * Sentiment analysis
  * News categorization
  * Document classification

* Uses word frequencies as features.

## Dataset: Simple Spam Detection Example

```python
texts = [
    "buy cheap medicine",
    "win a free prize",
    "meeting schedule confirmed",
    "project deadline tomorrow",
    "limited time offer",
    "let us discuss the report"
]

labels = [1,1,0,0,1,0]

vectorizer = CountVectorizer()

X_text = vectorizer.fit_transform(texts)

run_model(
    X_text,
    labels,
    test_size=0.33,
    model_type=MultinomialNB
)
```

# Bernoulli Naive Bayes (BNB)
* Designed for binary features.
* Each feature is either:

  * 1 (present)
  * 0 (absent)

Examples:

* Word present/not present
* Clicked/not clicked
* Purchased/not purchased
* Yes/No attributes

## Dataset: Artificial Binary Feature Dataset

```python
X, y = make_classification(
    n_samples=200,
    n_features=10,
    n_informative=5,
    n_redundant=0,
    random_state=42
)

X = (X > 0).astype(int)

run_model(
    X,
    y,
    test_size=0.25,
    model_type=BernoulliNB
)
```

# Conclusion

In this notebook, we implemented and compared three Naive Bayes classifiers:

| Model         | Best For                  |
| ------------- | ------------------------- |
| GaussianNB    | Continuous numerical data |
| MultinomialNB | Count-based text features |
| BernoulliNB   | Binary features (0/1)     |

Although all three are based on Bayes' Theorem, each variant is optimized for a different type of data representation. Choosing the correct variant significantly impacts model performance.
