# Linear Regression Using Scikit-Learn

## Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Dataset

| X | Y |
|---|---|
| 1 | 3 |
| 3 | 5 |
| 4 | 7 |
| 5 | 9 |
| 7 | 11 |

### Why Double Brackets?
Scikit-Learn expects features in 2D format: (samples, features)

Meaning:
- 5 samples
- 1 feature

```python
X = np.array([[1],[3],[4],[5],[7]])
Y = np.array([3,5,7,9,11])
```

## Visualizing Data Before Training

```python
plt.figure(figsize=(6,4))
plt.scatter(X,Y,color='blue')

plt.title("Before Linear Regression Using Scikit-Learn")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.savefig("Before Linear Regression Using Scikit-Learn.png", dpi=100)

plt.show()
```

## Linear Regression Model
Creates an instance of the Linear Regression model.

The model will automatically calculate:

- Best slope
- Best intercept

using the Least Squares Method.

```python
model = LinearRegression()
```

### Train the Model
Fits the model to the data.

The algorithm learns: `y = mx + b` by finding the values of `m` and `b` that minimize prediction error.

```python
model.fit(X,Y)
```

### Make Predictions
Uses the trained model to generate predicted values.

```python
y_pred = model.predict(X)
print("Predicted Values: ", y_pred)
```

## Visualizing Data After Training

```python
plt.figure(figsize=(6,4))

plt.scatter(X, Y, color='blue', label="Original Data")
plt.plot(X, y_pred, color='red', label="Best Fit Line")
plt.title("After Linear Regression Using Scikit-Learn")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.savefig(
    "After Linear Regression Using Scikit-Learn.png",
    dpi=100
)
plt.show()
```

## Extract Model Parameters

### Slope
- Returns the coefficient of the regression line.
- The slope tells how much Y changes when X increases by one unit.

```python
m = model.coef_
print("Slope (m): ", m)
```

### Intercept
Returns the point where the regression line crosses the Y-axis.
- When: x = 0
- Predicted value becomes: y = b

```python
b = model.intercept_
print("Y Intercept (b): ", b)
```
