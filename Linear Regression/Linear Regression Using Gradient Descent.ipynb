# Linear Regression Using Gradient Descent
Linear Regression tries to find the best straight line that fits the data.

The equation of a straight line is: y = mx + b

Where:

- m = Slope of the line
- b = Y-intercept
- x = Input value
- y = Predicted output

Gradient Descent is an optimization algorithm that repeatedly updates the values of (m) and (b) to reduce prediction error.

## Required Libraries

```python
import numpy as np
```

## Dataset
| X | Y |
|---|---|
| 1 | 3 |
| 2 | 4 |
| 3 | 2 |
| 4 | 5 |
| 5 | 6 |

- X represents input values.
- Y represents actual target values.
    

```python
# X = np.array([1,2,3,4,5])
X = np.array([1,3,4,5,7])
Y = np.array([3,5,7,9,11])
# Y = np.array([3,4,2,5,6])
```

## Initialize Parameters

- `Learning Rate`: Controls how much the parameters update during each iteration.
  - Small value = Slow learning
  - Large value = Fast learning but may overshoot
- `Epochs`: Number of complete training iterations.
- `m and b Initial Values`: Training starts with slope and intercept equal to zero.
- `Number of Samples (n)`: total number of data points.

```python
learning_rate = 0.01
epochs = 1000
m = 0.0
b = 0.0
n = len(X)
```

## Training Loop

### Cost Function
#### Mean Squared Error (MSE).
- Measures prediction error.
- Lower value means better model.

```python
for i in range(epochs):

    Y_pred = m * X + b      # Computes predicted values using the current line equation.

    cost = ((Y_pred - Y) ** 2).mean()

    derivative_m = (-2 / n) * sum(X * (Y - Y_pred)) # Gradient for Slope: Computes how much the slope should change.
    derivative_b = (-2 / n) * sum(Y - Y_pred)       # Gradient for Intercept: Computes how much the intercept should change.
   
    # Parameter Update
    m -= learning_rate * derivative_m
    b -= learning_rate * derivative_b

    print(f"Epoch {i+1}/{epochs}, Cost:{cost}, m:{m}, b:{b}")
```
