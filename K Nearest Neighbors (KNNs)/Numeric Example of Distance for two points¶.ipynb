### Numeric Example of Distance for two points

```python
import numpy as np
```

```python
x1=np.array([2,3])
x2=np.array([5,7])
```

```python
euclidean=np.linalg.norm(x1-x2)
manhattan=np.sum(np.abs(x1-x2))
minkowski=(np.sum(np.abs(x1-x2)**3))**(1/3)
```

```python
print(f"Distancce b/w {x1} and {x2}")
print(f"Euclidean {euclidean:.4f}")
print(f"Manhattan {manhattan:.4f}")
print(f"Minkowski (p=3) {minkowski:.3f}")
```
