# Hyperparameter Tuning and Evaluation using KNN Kernal (Euclidean, Manhattan, Minkowski) on Make Classification dataset

## Import Libraries

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
```

```python
import warnings
warnings.filterwarnings("ignore")
```

## Function

```python
def metrics(model,name,X_test):
    pred=model.predict(X_test)
    acc=accuracy_score(y_test,pred)
    cm=confusion_matrix(y_test, pred)
    print(f"\n-------- {name} --------\n")
    print(f"Accuracy: {acc:.4f}")
    print(f"Confusion Matrix:\n {cm}")
```

```python
def scalar_data(X_train,X_test):  
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
```

```python
def knn(name,metric,X_test,X_train,k,p=None):

    scalar_data(X_train,X_test)

    length = len(k) if isinstance(k, list) else 1
    if length>1:
        print("\n=============Hyperparameter Tunning============\n")
        best_acc=0
        best_k=None
        for k in k:
            if metric!="minkowski":
                model= KNeighborsClassifier(n_neighbors=k,metric=metric)
            else:
                model= KNeighborsClassifier(n_neighbors=k,metric=metric,p=p)
                
            model.fit(X_train,y_train)
            y_pred=model.predict(X_test)
            acc=accuracy_score(y_test,y_pred)
            print(f"k={k:>4}: Accuracy={acc:.4f}")
            if acc>best_acc:
                best_acc=acc
                best_k=k
            
        print(f"\nBest k: {best_k}")
        if metric!="minkowski":
            model= KNeighborsClassifier(n_neighbors=best_k,metric=metric)
        else:
           model= KNeighborsClassifier(n_neighbors=best_k,metric=metric,p=p) 
        
        model.fit(X_train,y_train) 
        metrics(model,name,X_test)

    else:
        if metric!="minkowski":
            model= KNeighborsClassifier(n_neighbors=k,metric=metric)
        else:
           model= KNeighborsClassifier(n_neighbors=k,metric=metric,p=p) 

        
        model.fit(X_train,y_train)
        metrics(model,name,X_test)
```

## Load and Train_Test Split Dataset

```python
X,y=make_classification(
    n_samples=1000,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    random_state=42
)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42
)
```

## Call

```python
knn("Euclidean","euclidean",X_test,X_train,list(range(1, 31, 2)))
```

```python
knn("Manhattan","manhattan",X_test,X_train,list(range(1, 31, 2)))
```
