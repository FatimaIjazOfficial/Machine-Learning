# Hyperparameter Tunning and Reduced Dimensionality using KNN on Breast Cancer Dataset

## Import Libraries

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

## Load and Split DataSet

```python
data = load_breast_cancer()

X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
print("Original Dimensions:", X_train.shape)
```

## Scalar Data

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Hyperparameter Tunning

```python
best_acc=0
best_k=None

for k in range(1,21):
    knn=KNeighborsClassifier(n_neighbors=k,metric="euclidean")
    knn.fit(X_train,y_train)
    y_pred=knn.predict(X_test)
    acc=accuracy_score(y_test,y_pred)

    print(f"k={k:>2}: Accuracy={acc:.4f}")

    if acc>best_acc:
        best_acc=acc
        best_k=k

print(f"Best k found: {best_k} with accuracy: {best_acc:.4f}")
```

## Reduced Dimensionality using PCA

```python
pca=PCA(n_components=10)
X_train_pca=pca.fit_transform(X_train)
X_test_pca=pca.transform(X_test)

print("Reduced Dimensions:", X_train_pca.shape)
```

```python
knn_pca=KNeighborsClassifier(n_neighbors=best_k)
knn_pca.fit(X_train_pca,y_train)
y_pred_pca=knn_pca.predict(X_test_pca)
acc_pca=accuracy_score(y_test,y_pred_pca)

print(f"Accuracy after PCA ={acc_pca:.4f}")
```

```python
variance_retained=round(np.sum(pca.explained_variance_ratio_),3)
print(f"Variance Retained: {variance_retained}")
```
