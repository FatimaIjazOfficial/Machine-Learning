# Random Forest Out-of-Bag Evaluation on Breast Cancer Dataset
- Random Forest is an ensemble machine learning algorithm that combines multiple Decision Trees to improve prediction accuracy and reduce overfitting.
- Instead of relying on a single tree, Random Forest builds many trees and combines their predictions.

**Out-of-Bag Evaluation**
- Each tree is trained using a bootstrap sample.
- Samples not selected for a particular tree are called Out-of-Bag samples.
- These samples can be used as a validation set without creating a separate validation dataset.

## Import Libraries

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

## Load DataSet

```python
data = load_breast_cancer()
```

```python
X = data.data
y = data.target
feature_names = data.feature_names
class_names = data.target_names
```

## Train-Test Split
Machine learning models should be evaluated on data they have never seen before.

If we train and test on the same data, the model may simply memorize the training examples, giving an unrealistic estimate of performance.

To avoid this issue, the dataset is divided into:

* Training Data
* Testing Data

The training set is used to learn patterns, while the testing set is used to evaluate how well the model generalizes to unseen samples.

Using `stratify=y` ensures that each class maintains approximately the same proportion in both training and testing datasets.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
```

## Build Random Forest Model

```python
rf = RandomForestClassifier(
    n_estimators=100,      # Number of decision trees.
    bootstrap=True,        # Required for Out-of-Bag evaluation.
    oob_score=True,        # Computes OOB accuracy using unused bootstrap samples.
    random_state=42,       # Ensures reproducible results.
    n_jobs=-1              # Uses all CPU cores.
)
```

## Train Model

```python
rf.fit(X_train, y_train)
```

## OOB Accuracy and Error

```python
oob_accuracy = rf.oob_score_
oob_error = 1 - oob_accuracy

print("-:Out of Bag Evaluation:-")
print(f"OOB Accuracy: {oob_accuracy:.4f}")
print(f"OOB Error: {oob_error:.4f}")
print("(No separate validation set was used)")
```
