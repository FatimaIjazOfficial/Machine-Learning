# Make BLOBS Analysis Using Hierarchical Clustering (Agglomerative, Divisive, FCLuster and Dendrogram) with Scalar 

## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering, BisectingKMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="joblib")

import os
os.environ["LOKY_MAX_CPU_COUNT"] = "1"
```

## Generate synthetic dataset

```python
X_blob, y_true = make_blobs(
    n_samples=200,
    centers=3,
    cluster_std=0.9,
    random_state=42
)

# Convert to DataFrame
X = pd.DataFrame(
    X_blob,
    columns=["Feature_1", "Feature_2"]
)

print("Dataset Shape:", X.shape)
print("\nFirst 5 Records:")
X.head()
```

## Standardize features (Before clustering)

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dont need because it is already 2 Dimensional
X_pca = X_scaled
```

## Agglomerative Clustering (Bottom-Up)

```python
agg = AgglomerativeClustering(
    n_clusters=3,
    linkage="ward"
)
```

## Bisecting K-Means (Top-Down)

```python
div = BisectingKMeans(
    n_clusters=3,
    random_state=42
)
```

## Linkage matrix

```python
Z = linkage(
    X_scaled,
    method="ward"
)
```

## Cluster labels

```python
agg_labels = agg.fit_predict(X_scaled)
div_labels = div.fit_predict(X_scaled)
dendro_labels = fcluster(Z, t=3, criterion="maxclust") # Extract 3 clusters from the dendrogram

X["Agglomerative_Cluster"] = agg_labels
X["Bisecting_Cluster"] = div_labels
X["Dendrogram_Cluster"] = dendro_labels
```

### Sample records

```python
print("\nDataset with Cluster Labels:")
X.head()
```

### Cluster counts

```python
print("\nAgglomerative Cluster Counts")

for cluster, count in X["Agglomerative_Cluster"].value_counts().sort_index().items():
    print(f"Cluster {cluster}: {count} samples")

print("\nBisecting K-Means Cluster Counts")

for cluster, count in X["Bisecting_Cluster"].value_counts().sort_index().items():
    print(f"Cluster {cluster}: {count} samples")

print("\nDendrogram Cluster Counts")

for cluster, count in X["Dendrogram_Cluster"].value_counts().sort_index().items():
    print(f"Cluster {cluster}: {count} samples")    
```

## Visualization

### Dendrogram

```python
plt.figure(figsize=(10, 5))

dendrogram(Z)

plt.title("Blobs Hierarchical Clustering Dendrogram")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.savefig("Blobs Hierarchical Clustering Dendrogram.png",dpi=100)
plt.show()
```

### Clustering methods

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))


axes[0].scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    c=agg_labels,
    cmap="tab10",
    s=30
)

axes[0].set_title("Agglomerative Clustering")
axes[0].set_xlabel("PC1")
axes[0].set_ylabel("PC2")


axes[1].scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    c=div_labels,
    cmap="tab10",
    s=30
)

axes[1].set_title("Bisecting K-Means")
axes[1].set_xlabel("PC1")
axes[1].set_ylabel("PC2")


axes[2].scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    c=dendro_labels,
    cmap="tab10",
    s=30
)

axes[2].set_title("Clusters from Dendrogram")
axes[2].set_xlabel("PC1")
axes[2].set_ylabel("PC2")

plt.savefig("Blobs Clustring Methods.png",dpi=100)
plt.tight_layout()
plt.show()
```
