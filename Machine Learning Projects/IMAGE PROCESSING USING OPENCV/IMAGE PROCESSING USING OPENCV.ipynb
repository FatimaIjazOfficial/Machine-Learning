# IMAGE PROCESSING USING OPENCV

## IMPORT LIBRARIES

```python
import cv2 # OpenCV library for image processing
import numpy as np # numerical operations and array handling
from google.colab.patches import cv2_imshow # display images in Google Colab
```

## LOAD IMAGE
- Reads image from storage into memory
- Image is loaded as a NumPy array
- Default format in OpenCV: BGR (Blue, Green, Red)

```python
image = cv2.imread("sunflower.jpg")

if image is None:
    print("Error: Image not found")
    exit()

print("Image Shape:", image.shape)

cv2_imshow(image)
cv2.waitKey(0)
```

## GRAYSCALE CONVERSION

Color images contain 3 channels (B, G, R)

Grayscale image contains only intensity values:
- 0   → black
- 255 → white

Benefits:
- Reduces computational cost
- Removes unnecessary color information
- Useful for edge detection and feature extraction

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2_imshow(gray)
cv2.waitKey(0)
```

## GAUSSIAN BLUR
Gaussian Blur smooths the image and reduces noise

Idea:
- Each pixel becomes weighted average of neighboring pixels
- Gaussian kernel gives more importance to center pixels

Benefits:
- Removes small image noise
- Improves edge detection quality

```python
blurred = cv2.GaussianBlur(
    gray,
    (5,5),   # kernel size
    0        # sigma automatically calculated
)

cv2_imshow(blurred)
cv2.waitKey(0)
```

## MEDIAN BLUR

Median Blur replaces each pixel with median value of neighboring pixels

Benefits:
- Removes salt-and-pepper noise
- Preserves edges better than average blur

```python
median = cv2.medianBlur(
    gray,
    5
)

cv2_imshow(median)
cv2.waitKey(0)
```

## EDGE DETECTION USING CANNY

Edge detection finds object boundaries

Main idea:
1. Smooth image
2. Calculate intensity gradients
3. Keep strong edges
4. Remove weak edges

Thresholds:
- 100 → lower threshold
- 200 → upper threshold

```python
edges = cv2.Canny(
    blurred,
    100,
    200
)

cv2_imshow(edges)
cv2.waitKey(0)
```

## IMAGE RESIZING

Resizing changes image dimensions

Benefits:
- Standardizes image size
- Reduces memory usage
- Speeds up processing

New size:
- Width  = 400 pixels
- Height = 300 pixels

```python
resized = cv2.resize(image, (400,300))

cv2_imshow(resized)
cv2.waitKey(0)
```

## RESIZE GRAYSCALE IMAGE

```python
resized_gray = cv2.resize(
    gray,
    (400,300)
)

cv2_imshow(resized_gray)
cv2.waitKey(0)
```

## RESIZE GAUSSIAN BLUR IMAGE

```python
resized_blurred = cv2.resize(
    blurred,
    (400,300)
)

cv2_imshow(resized_blurred)
cv2.waitKey(0)
```

## RESIZE MEDIAN BLUR IMAGE

```python
resized_median = cv2.resize(
    median,
    (400,300)
)

cv2_imshow(resized_median)
cv2.waitKey(0)
```

## RESIZE EDGE IMAGE

```python
resized_edges = cv2.resize(
    edges,
    (400,300)
)

cv2_imshow(resized_edges)
cv2.waitKey(0)
```

## IMAGE ROTATION

Rotation changes image orientation around a center point

Center:
- x = width / 2
- y = height / 2

Parameters:
- Angle = 45°
- Scale = 1 (original size)

```python
height, width = image.shape[:2]

center = (
    width / 2,
    height / 2
)

rotation_matrix = cv2.getRotationMatrix2D(
    center,
    45,
    1
)

rotated_image = cv2.warpAffine(
    image,
    rotation_matrix,
    (width, height)
)

cv2_imshow(rotated_image)
cv2.waitKey(0)
```

## IMAGE FLIPPING

Flipping creates mirror image

flipCode:
- 0  → vertical flip
- 1  → horizontal flip
- -1 → both directions

```python
flipped = cv2.flip(
    image,
    1
)

cv2_imshow(flipped)
cv2.waitKey(0)
```

## SAVE OUTPUT IMAGE

Stores processed image on disk

Here:
- Edge-detected image is saved
- File name: edges_output.jpg

```python
cv2.imwrite(
    "edges_output.jpg",
    edges
)
```

## CLOSE WINDOWS

Releases OpenCV resources and closes windows

```python
cv2.destroyAllWindows()
```
