In this notebook, we:

This notebook provides the foundation for understanding how Decision Trees make classification decisions before moving to overfitting and pruning techniques.

## Decision Tree
A Decision Tree is a supervised machine learning algorithm used for:

* Classification
* Regression

It learns a series of decision rules that split data into smaller groups until samples belonging to the same class are grouped together.

### Example

```text
Is Feature1 > 0.5 ?

├── Yes → Class 1
└── No
     ├── Feature2 > -1 ?
     │      ├── Yes → Class 0
     │      └── No → Class 1
```

Each question is called a **split**.

The final nodes are called **leaf nodes** and contain the predicted class.

## Imports
| Library                          | Purpose                                      |
| -------------------------------- | -------------------------------------------- |
| NumPy (`numpy`)                  | Numerical computations and array operations  |
| Matplotlib (`matplotlib.pyplot`) | Data visualization and plotting              |
| `DecisionTreeClassifier`         | Builds a Decision Tree classification model  |
| `export_text`                    | Displays tree rules in text format           |
| `make_classification`            | Generates a synthetic classification dataset |

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.datasets import make_classification
````

## Generate Dataset
| Variable | Description                 |
| -------- | --------------------------- |
| `X`      | Feature matrix (input data) |
| `y`      | Target labels (classes)     |

````python
X, y = make_classification(
    n_samples=12,                    # Total number of observations
    n_features=2,                    # Total input features
    n_informative=2,                 # Features that contain useful information
    n_redundant=0,                   # Duplicate or unnecessary features
    random_state=42                  # Produces the same dataset every run
)
````

## Display Dataset

| Component       | Purpose                               |
| --------------- | ------------------------------------- |
| `feature_names` | User-friendly feature labels          |
| `X[i,0]`        | Value of Feature1                     |
| `X[i,1]`        | Value of Feature2                     |
| `y[i]`          | Corresponding class label             |
| `.2f`           | Displays values with 2 decimal places |
    
Each row represents one sample containing:

```text
Features → Prediction Target
```
    

````python
feature_names = ["Feature1", "Feature2"]

print("Dataset:")

for i in range(len(y)):
    print(
        f"{feature_names[0]}={X[i,0]:.2f}, "
        f"{feature_names[1]}={X[i,1]:.2f}, "
        f"Class={y[i]}"
    )
````

## Train Decision Tree

### Create Model

#### Gini Impurity

Gini Impurity measures how mixed the classes are inside a node.

| Gini Value | Meaning              |
| ---------- | -------------------- |
| 0          | Completely pure node |
| Close to 1 | Highly mixed node    |

Lower values are preferred because they produce cleaner class separation.

````python
tree = DecisionTreeClassifier(
    max_depth=2,                     #  Maximum allowed tree depth
    criterion="gini",                # Splitting method
    random_state=42                  # Reproducible results
)
````

### Train Model
During training, the tree learns the best feature splits that separate the classes.
| Component | Purpose          |
| --------- | ---------------- |
| `fit()`   | Trains the model |
| `X`       | Input features   |
| `y`       | Target labels    |
    

````python
tree.fit(X, y)
````

## Tree Structure

### Why `export_text()`?

Instead of displaying a graphical tree, it prints the learned decision rules in text format.

````python
print(
    export_text(
        tree,
        feature_names=feature_names
    )
)
````

| Rule               | Meaning            |
| ------------------ | ------------------ |
| `Feature2 <= -0.19` | Go to left branch  |
| `Feature2 >  -0.19`  | Go to right branch |
| `Feature1 <= 0.78` | Go to left branch  |
| `Feature1 >  0.78`  | Go to right branch |
| `class: 0`         | Predict Class 0    |
| `class: 1`         | Predict Class 1    |

The model uses these rules to classify unseen samples.

## Feature Importance
Feature Importance shows how much each feature contributed to the model's decisions.

````python
for name, importance in zip(
    feature_names,
    tree.feature_importances_
):
    print(f"{name}: {importance:.3f}")
````

| Feature  | Importance | Meaning                |
| -------- | ---------- | ---------------------- |
| Feature1 | 0.082      | Less important feature |
| Feature2 | 0.918      | Most important feature |

Higher importance means the feature was used more frequently or provided more useful splits during training.
